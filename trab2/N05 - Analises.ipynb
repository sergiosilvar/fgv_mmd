### Some analysis using pandas.

```python
import pandas.io.sql as psql
import sqlite3
import numpy as np
import pandas
from pandas import Series
%pylab inline

def dataframe(where_clause='', index_col=None):
    # Concectar ao banco de dados e criar um DataFrame.
    con = sqlite3.connect('zap.db')
    sql = 'select * from vw_home '
    df = psql.read_frame( sql + where_clause, index_col=index_col,con=con)
    con.close()
    return df
```

```python
# Overview of data available.
allrows = dataframe()


# Filter homes within Rio de Janeiro city limits.
north = -22.782825
south = -23.075116
east = -43.160987
west = -43.716692

rj = allrows[allrows.lat_home >= south]
rj = rj[rj.lat_home <= north]
rj = rj[rj.lng_home >= west]
rj = rj[rj.lng_home <= east]
```

```python
# After the filter.
rj.describe()
```

```python
pylab.rcParams['figure.figsize'] = (10.0, 8)
```

```python
df = rj
pylab.plot(df.lat_home, df.price_m2, '.')
pylab.title(u'Preço do m² com base na latitude.')
pylab.xlabel('Latitude')
pylab.ylabel(u'R$ 1.000/m²')
```

```python
# Plt m^2 price based on longitude.
pylab.plot(df.lng_home, df.price_m2, '.')
pylab.title(u'Preço do m² com base na longitude.')
pylab.xlabel('Longitude')
pylab.ylabel(u'R$ 1.000/m²')
```

```python
# Scatter plot de preço do m^2 para um determinaod conjunto de apartamentos usanco Colormap e coordenadas como eixos.
x = df[df.bedrooms==1]
# Colormaps: http://wiki.scipy.org/Cookbook/Matplotlib/Show_colormaps
pylab.scatter(y=x.lat_home,x=x.lng_home, s=60, c=x.price_m2, cmap=pylab.get_cmap('BuPu'))
pylab.ylim(-23.1,-22.7)
pylab.xlim(-43.7,-43.1)
```

```python
# Home wiht R$/m² the most.
df.loc[df.price_m2.idxmax()]['url']
```

```python
m2 =  pandas.DataFrame(data= df.groupby('neigh').price_m2.mean())
m2['glat']=df.groupby('neigh').glat.mean()
m2['glng']=df.groupby('neigh').glng.mean()
m2 = m2.sort(ascending=False, column='price_m2')
m2
```

```python
# Export to CSV for further processing in other tools
m2.to_csv('price_m2.csv',encoding='utf-8')
```
